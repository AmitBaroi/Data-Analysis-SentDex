```python
import datetime
# pandas.io.data as web has been depricated, now we use pandas_datareader.data:
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
style.use('ggplot')
```

We will get stock data from 1st January 2010 to today

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime.now()

# Getting stock data from a web API
df = web.DataReader("XOM", "yahoo", start, end)
df2 = web.DataReader("AAPL", "yahoo", start, end)
```

We will index our data by Date

```python
# Resetting index to clear any unwanted index columns
df.reset_index(inplace=True)
# We will index by Date
df.set_index("Date", inplace=True)
# Showing the first 5 rows of data
df.head()
```

Setting index for second data

```python
# Resetting index to clear any unwanted index columns
df2.reset_index(inplace=True)
# We will index by Date
df2.set_index("Date", inplace=True)
# Showing the first 5 rows of data
df2.head()
```

Opening stock prices

```python
plt.figure(figsize=(16, 5))
plt.title('Comparison of opening stock prices for Exxon-Mobil (XOM) and Apple (AAPL)')
sns.lineplot(data=df['Open'], label='XOM')
sns.lineplot(data=df2['Open'], label='AAPL')
plt.legend()
```

We can see that the opening and closing prices overlap quite a bit and follow a similar trend.
